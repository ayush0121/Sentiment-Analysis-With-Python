```python
import requests  #to get the data of the yelp (buisnesss directories)
```

```python
from bs4 import BeautifulSoup
```

```python
#create a requests that contains our review data
r=requests.get('https://www.yelp.com/biz/westfield-san-francisco-centre-san-francisco-3')
```

```python
r.status_code
```

```python
r.text
```

```python
soup=BeautifulSoup(r.text,'html.parser')
```

```python
divs=soup.findAll(class_=' raw__09f24__T4Ezm')
```

```python
reviews=[]
for div in divs:
    reviews.append(div.find('p').text)
```
